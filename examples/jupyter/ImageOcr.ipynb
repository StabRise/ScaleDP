```python
from sparkpdf import *

spark = SparkPdfSession()
```

```python
image_example = "../../sparkpdf/resources/images/Invoice.png"
df = spark.read.format("binaryFile") \
    .load([image_example]*2)

#df.show_image("content")
```

```python
pipeline = PipelineModel(stages=[
    DataToImage(inputCol="content", outputCol="image"),
    EasyOcr(inputCol="image", outputCol="text", partitionMap=False, numPartitions=1, device=Device.CPU, batchSize=5, keepFormatting=True),
])

result = pipeline.transform(df).cache()

result.show_text()
```

```python

pipeline = PipelineModel(stages=[
    DataToImage(inputCol="content", outputCol="image"),
    SuryaOcr(inputCol="image", outputCol="text", partitionMap=True, numPartitions=1, device=Device.CUDA, batchSize=5, keepFormatting=True),
])

result = pipeline.transform(df).cache()

result.show_text()
```
